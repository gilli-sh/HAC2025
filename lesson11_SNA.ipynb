```python
import networkx as nx
import matplotlib.pyplot as plt

# יצירת גרף לא מכוון
G = nx.Graph()

# הוספת צמתים וקשתות
G.add_edges_from([("Alice", "Bob"), ("Bob", "Charlie"), ("Alice", "Charlie")])

# ציור הגרף
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color="skyblue", node_size=3000, font_size=15)
plt.title("רשת חברתית פשוטה")
plt.show()

# חישוב Degree של כל צומת
degrees = dict(G.degree())
print("Degree של כל צומת:", degrees)
```

```python
# דוגמה 
# יצירת גרף מכוון
G = nx.DiGraph()
    
# הוספת צמתים וקשתות    
G.add_edges_from([("Alice", "Bob"), ("Bob", "Charlie"), ("Charlie", "Alice")])

# ציור הגרף
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color="skyblue", node_size=3000, font_size=15)
plt.title("רשת חברתית פשוטה")
plt.show()
    
# חישוב Degree של כל צומת   
degrees = dict(G.degree())
print("Degree של כל צומת:", degrees)
```
